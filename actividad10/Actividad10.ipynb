```python
from math import cos
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import ode
import pandas as pd


w = []
graf = []
der=[]

        
#Definimos las ecuaciones diferenciales y los parámetros
def f(t, z, p):
    """"
        Lado derecho de las ecuaciones diferenciales
        dx/dt = y
        dy/dt = - δ*y - α*x - β*x**3 + γ*cos(ω*t)

    """
    δ, α, β, γ, ω = p                           #Parámetros de nuestra integral
    x, y = z                                    #Parámetros de integración
    f = [y, - δ*y - α*x - β*x**3 + γ*cos(ω*t)]  #Función a integrar

    return f

```

```python
def derivada(t,x1,x2):
  
    derivada= (x1 - x2)/(1*t)   

    return derivada
```

```python
solver = ode(f)
solver.set_integrator('dopri5')

β1j = -0.003
wi1=[]
grafi1=[]
deri1=[]

for i in range(0,251):
    γ = 1.0
    α = 1.0        
    δ = 0.1
    ω = i*0.01
    β = β1j
    val = δ, α, β, γ, ω                      
    solver.set_f_params(val)

    t0 = 0.0
    z0 = [1.0, 0]
    solver.set_initial_value(z0, t0)
    t1 = 40
    n = 400
    t = np.linspace(t0, t1, n)
    sol = np.empty((n, 2))
    sol[0] = z0
    k = 1

    while solver.successful() and solver.t < t1:
        solver.integrate(t[k])
        sol[k] = solver.y
        k += 1

    x = sol[:,0]

    wi1.append(ω)
    grafi1.append(max(abs(x)))

 

```
